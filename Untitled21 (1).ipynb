```python
#Assignment

#Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
#Answer: Anomaly Detection is the process of identifying data points that are significantly different from normal data patterns.
# Types of Anomalies:
#1️⃣ Point Anomaly
#A single data point is abnormal.
#Example: A ₹5,00,000 transaction when usual spending is ₹5,000.
#2️⃣ Contextual Anomaly
#A data point is abnormal in a specific context (time, season, location).
#Example: 40°C temperature in winter.
#3️⃣ Collective Anomaly
#A group of data points together form an abnormal pattern.
#Example: Continuous unusual network traffic indicating a cyber attack.

#Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
#Answer: Isolation Forest → Uses random trees to isolate anomalies.
# Best for large and high-dimensional datasets (e.g., fraud detection).
#DBSCAN → Density-based clustering; low-density points are anomalies.
# Best for spatial data and datasets with noise.
#Local Outlier Factor (LOF) → Compares local density of a point with its neighbors.
# Best for detecting local outliers in varying-density data.

#Question 3: What are the key components of a Time Series? Explain each with one example.
#Answer: Trend – Long-term increase or decrease in data.
#Example: Sales increasing over years.
# Seasonality – Regular pattern that repeats at fixed intervals.
#Example: High sales during festivals.
#️ Cyclical – Long-term fluctuations without fixed period.
#Example: Economic boom and recession.
#️Irregular – Random, unpredictable variations.
#Example: Sudden drop due to natural disaster.

#Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?
#Answer: A time series is stationary if its statistical properties such as mean, variance, and autocorrelation remain constant over time.
#Types:
#Strict Stationary – Full distribution does not change over time.
#Weak (Covariance) Stationary – Mean and variance are constant, and covariance depends only on lag.
#How to Test Stationarity:
#Visual Inspection – Check for trend or seasonality in the plot.
#ADF Test (Augmented Dickey-Fuller Test) –
#.Null hypothesis: Series is non-stationary.
#.If p-value < 0.05 → series is stationary.
#KPSS Test – Confirms stationarity.
#How to Make Non-Stationary Series Stationary:
#Log / Square Root Transformation – Stabilizes variance.
#Removing Trend – Using regression or moving average.#Seasonal Differencing – Remove seasonal effects.

#Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
#Answer: AR – Uses past values of the series. (For stationary data).
#MA – Uses past error terms. (For shock-based patterns).
#ARIMA (p,d,q) – AR + MA + differencing. (Handles non-stationary data).
#SARIMA (p,d,q)(P,D,Q)s – ARIMA + seasonal components. (Handles seasonality).
#SARIMAX – SARIMA + external variables. (Includes outside factors).





```

```python
#Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load dataset
df = pd.read_csv(r"C:\Users\pc\OneDrive\Desktop\AirPassengers.csv", parse_dates=['Month'], index_col='Month')

# Plot original series
plt.figure(figsize=(12,6))
plt.plot(df['#Passengers'], label='AirPassengers')
plt.title("Monthly AirPassengers Data (1949-1960)")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(df['#Passengers'], model='multiplicative', period=12)

# Plot decomposition
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.show()
```

```python
#Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Load dataset (change path if needed)
df = pd.read_csv("C:/Users/pc/OneDrive/Desktop/NYC_taxi_fare_data.csv")

# Select only numerical columns
numeric_df = df.select_dtypes(include=[np.number])

# Drop missing values
numeric_df = numeric_df.dropna()

# For visualization, select two important features
# (example: fare_amount and trip_distance if available)
X = numeric_df[['fare_amount', 'trip_distance']]

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply Isolation Forest
iso_forest = IsolationForest(contamination=0.05, random_state=42)
y_pred = iso_forest.fit_predict(X_scaled)

# Add anomaly column
X['anomaly'] = y_pred

# Plot results
plt.figure()
normal = X[X['anomaly'] == 1]
anomaly = X[X['anomaly'] == -1]

plt.scatter(normal['fare_amount'], normal['trip_distance'])
plt.scatter(anomaly['fare_amount'], anomaly['trip_distance'])

plt.xlabel("Fare Amount")
plt.ylabel("Trip Distance")
plt.title("Isolation Forest - NYC Taxi Fare Anomaly Detection")
plt.show()
```

```python
#Question 8: Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load dataset
df = pd.read_csv("C:/Users/pc/OneDrive/Desktop/AirPassengers.csv")

# Rename columns if needed
df.columns = ['Month', 'Passengers']

# Convert Month to datetime
df['Month'] = pd.to_datetime(df['Month'])

# Set index
df.set_index('Month', inplace=True)

# Plot original data
plt.figure()
plt.plot(df['Passengers'])
plt.title("Monthly Airline Passengers")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()

# Train SARIMA Model
# (p,d,q) = (1,1,1)
# (P,D,Q,s) = (1,1,1,12)  -> seasonal period = 12 (monthly data)

model = SARIMAX(df['Passengers'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12))

results = model.fit()

# Forecast next 12 months
forecast = results.forecast(steps=12)

# Plot forecast
plt.figure()
plt.plot(df['Passengers'])
plt.plot(forecast)
plt.title("SARIMA Forecast - Next 12 Months")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.show()

print("Forecasted Values for Next 12 Months:")
print(forecast)
```

```python
#Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs

# Generate synthetic numerical dataset
X, _ = make_blobs(n_samples=300, centers=1, cluster_std=1.0, random_state=42)

# Add some artificial outliers
rng = np.random.RandomState(42)
outliers = rng.uniform(low=-6, high=6, size=(20, 2))
X = np.vstack([X, outliers])

# Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.06)
y_pred = lof.fit_predict(X)

# Separate normal and anomaly points
normal = X[y_pred == 1]
anomaly = X[y_pred == -1]
plt.figure()
plt.scatter(normal[:, 0], normal[:, 1])
plt.scatter(anomaly[:, 0], anomaly[:, 1])
plt.title("Local Outlier Factor (LOF) - Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()


```

```python
## Plot resultsQuestion 10: You are working as a data scientist for a power grid monitoring company.Your goal is to forecast energy demand and also detect abnormal spikes or drops inreal-time consumption data collected every 15 minutes. The dataset includes featureslike timestamp, region, weather conditions, and energy usage.
#Explain your real-time data science workflow:
#● How would you detect anomalies in this streaming data (Isolation Forest / LOF /DBSCAN)?
#● Which time series model would you use for short-term forecasting (ARIMA /SARIMA / SARIMAX)?
#● How would you validate and monitor the performance over time?
#● How would this solution help business decisions or operations?

#Answer-Use Isolation Forest because:
#.Fast and scalable for streaming data.
#.Detects sudden spikes/drops.

#Use SARIMAX because:
#.Handles trend + seasonality.
#.Can include external variables (weather).

#Validation & Monitoring:
#.Use MAE / RMSE for forecasting accuracy.
#.Monitor false anomaly rate.

#Business Impact:
#.Prevent blackouts.
#.Detect equipment failure early.

# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_error

# -----------------------------
# Simulate 15-minute energy data
# -----------------------------

np.random.seed(42)
date_range = pd.date_range(start="2024-01-01", periods=1000, freq="15T")

energy_usage = 200 + 20*np.sin(np.arange(1000)/50) + np.random.normal(0,5,1000)

df = pd.DataFrame({
    "timestamp": date_range,
    "energy_usage": energy_usage
})

df.set_index("timestamp", inplace=True)

# -----------------------------
# 1️⃣ Anomaly Detection
# -----------------------------

iso = IsolationForest(contamination=0.02, random_state=42)
df["anomaly"] = iso.fit_predict(df[["energy_usage"]])

# Plot anomalies
plt.figure()
normal = df[df["anomaly"] == 1]
anomaly = df[df["anomaly"] == -1]

plt.scatter(normal.index, normal["energy_usage"])
plt.scatter(anomaly.index, anomaly["energy_usage"])
plt.title("Real-Time Energy Anomaly Detection")
plt.show()

# -----------------------------
# 2️⃣ Forecasting using SARIMAX
# -----------------------------

model = SARIMAX(df["energy_usage"],
                order=(1,1,1),
                seasonal_order=(1,1,1,96))  # 96 = 15min * 96 = 1 day

results = model.fit(disp=False)

forecast = results.forecast(steps=96)  # next 1 day

# Plot forecast
plt.figure()
plt.plot(df["energy_usage"][-200:])
plt.plot(forecast)
plt.title("Short-Term Energy Forecast (Next 24 Hours)")
plt.show()

# -----------------------------
# 3️⃣ Validation
# -----------------------------

train = df["energy_usage"][:-96]
test = df["energy_usage"][-96:]

model_val = SARIMAX(train,
                    order=(1,1,1),
                    seasonal_order=(1,1,1,96)).fit(disp=False)

pred = model_val.forecast(steps=96)

mae = mean_absolute_error(test, pred)
print("Validation MAE:", mae)

```
